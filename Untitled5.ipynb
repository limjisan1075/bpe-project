```python
import os
import glob
import numpy as np
import pandas as pd
import cv2
import pywt # pip install PyWavelets 필요
from skimage.feature import graycomatrix, graycoprops # pip install scikit-image 필요

# ==========================================================
# 1. 수학적 BPE 정량화 핵심 파이프라인 함수
# ==========================================================
def calculate_mathematical_bpe(image_path):
    # 이미지 로드 (Subtracted/Recombined 이미지 기준)
    img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    if img is None:
        return None

    # ------------------------------------------------------
    # STEP 1: 해부학적 ROI 분리 (배경/피부/대흉근 및 FGT 마스킹)
    # ------------------------------------------------------
    # 유방 전체 마스크 (Otsu Thresholding)
    _, breast_mask = cv2.threshold(img, 15, 255, cv2.THRESH_BINARY)

    # 지방(Fat) 영역 기준 신호 통계 (유방 내부 중 하위 30% 신호 강도)
    breast_pixels = img[breast_mask > 0]
    if len(breast_pixels) == 0:
        return None

    fat_threshold = np.percentile(breast_pixels, 30)
    fat_pixels = breast_pixels[breast_pixels <= fat_threshold]
    mu_fat = np.mean(fat_pixels)
    sigma_fat = np.std(fat_pixels) + 1e-5 # Zero division 방지

    # FGT (실질 조직) 후보 영역 (상위 70% 신호)
    fgt_mask = (img > fat_threshold) & (breast_mask > 0)

    # ------------------------------------------------------
    # STEP 2: 국소 고신호 종괴(Mass) 수학적 억제 (Morphological Top-Hat)
    # ------------------------------------------------------
    # 31x31 커널로 덩어리진 대형 종괴(Low-frequency Peak) 추출 후 차감
    kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (31, 31))
    top_hat = cv2.morphologyEx(img, cv2.MORPH_TOPHAT, kernel)

    # ------------------------------------------------------
    # STEP 3: 수학적 정량 지표 산출
    # ------------------------------------------------------
    # A. Z-score 기반 조영증강 강도 (Contrast Score)
    z_score_map = (img.astype(np.float32) - mu_fat) / sigma_fat
    fgt_z_scores = z_score_map[fgt_mask]

    mean_z_score = np.mean(fgt_z_scores) if len(fgt_z_scores) > 0 else 0
    high_bpe_ratio = np.sum(fgt_z_scores > 2.0) / (np.sum(fgt_mask) + 1e-5) # Z > 2.0 비율

    # B. 웨이브렛(Wavelet) 고주파 분해 (거미줄 텍스처 패턴 에너지)
    coeffs = pywt.dwt2(top_hat, 'db1')
    cA, (cH, cV, cD) = coeffs
    # Detail 계수들의 L2 Norm으로 거미줄 텍스처 에너지 측정
    wavelet_texture_energy = np.mean(np.sqrt(cH**2 + cV**2 + cD**2))

    # C. GLCM 공간 엔트로피 (Spatial Entropy & Homogeneity)
    # 연산 속도를 위해 16-level 양자화
    img_quantized = (top_hat / 16).astype(np.uint8)
    glcm = graycomatrix(img_quantized, distances=[1], angles=[0, np.pi/4, np.pi/2], levels=16, symmetric=True, normed=True)
    glcm_contrast = np.mean(graycoprops(glcm, 'contrast'))
    glcm_homogeneity = np.mean(graycoprops(glcm, 'homogeneity'))

    # ------------------------------------------------------
    # STEP 4: 최종 통합 정량 BPE Score (0~100 수치화)
    # ------------------------------------------------------
    # 조영증강 면적비율 + Z-score + 거미줄 텍스처 에너지를 결합한 연속형 수치
    raw_bpe_score = (high_bpe_ratio * 40.0) + (mean_z_score * 10.0) + (wavelet_texture_energy * 0.5)
    bpe_score_continuous = np.clip(raw_bpe_score, 0.0, 100.0)

    # 연속형 점수를 바탕으로 4단계 표준 범주형 라벨 할당 (Quantile/Threshold)
    if bpe_score_continuous < 15.0:
        bpe_class_4 = 0 # Minimal
    elif bpe_score_continuous < 35.0:
        bpe_class_4 = 1 # Mild
    elif bpe_score_continuous < 65.0:
        bpe_class_4 = 2 # Moderate
    else:
        bpe_class_4 = 3 # Marked

    return {
        'Z_Score_Mean': round(float(mean_z_score), 4),
        'High_Enhancement_Ratio': round(float(high_bpe_ratio), 4),
        'Wavelet_Texture_Energy': round(float(wavelet_texture_energy), 4),
        'GLCM_Contrast': round(float(glcm_contrast), 4),
        'GLCM_Homogeneity': round(float(glcm_homogeneity), 4),
        'Math_BPE_Score_Continuous': round(float(bpe_score_continuous), 2),
        'Math_BPE_Class_4': bpe_class_4 # 0:Minimal, 1:Mild, 2:Moderate, 3:Marked
    }

# ==========================================================
# 2. 326명 환자 일괄 배치 처리 및 CSV 생성
# ==========================================================
def build_patient_bpe_csv(dataset_dir, output_csv_path):
    print("🚀 326명 환자 데이터 수학적 BPE 정량화 시작...")

    image_paths = sorted(glob.glob(os.path.join(dataset_dir, "*.png")))
    results = []

    for img_path in image_paths:
        fname = os.path.basename(img_path)
        # 환자 ID 추출 (예: P311_L_CM_CC_padded.png -> P311)
        patient_id = fname.split('_')[0]
        view_type = 'CC' if 'CC' in fname else ('MLO' if 'MLO' in fname else 'UNKNOWN')

        bpe_metrics = calculate_mathematical_bpe(img_path)
        if bpe_metrics is not None:
            bpe_metrics['Filename'] = fname
            bpe_metrics['Patient_ID'] = patient_id
            bpe_metrics['View_Type'] = view_type
            results.append(bpe_metrics)

    df = pd.DataFrame(results)

    # 컬럼 순서 정돈
    columns_order = [
        'Patient_ID', 'Filename', 'View_Type',
        'Math_BPE_Score_Continuous', 'Math_BPE_Class_4',
        'High_Enhancement_Ratio', 'Z_Score_Mean',
        'Wavelet_Texture_Energy', 'GLCM_Contrast', 'GLCM_Homogeneity'
    ]
    df = df[columns_order]

    df.to_csv(output_csv_path, index=False, encoding='utf-8-sig')
    print(f"✅ CSV 생성 완료: {output_csv_path}")
    print(f"📊 총 {len(df)}개 이미지에 대한 정량 지표 추출 성공!")
    print(df.head())

# 실행 (경로 수정 후 사용)
dataset_dir = "/content/drive/MyDrive/CDD-CESM/Subtracted_padded_512"
output_csv = "/content/drive/MyDrive/Mathematical_BPE_326_Patients.csv"
build_patient_bpe_csv(dataset_dir, output_csv)
```

```python
import pandas as pd
import numpy as np

# 1. 기존 CSV 읽기
csv_path = '/content/drive/MyDrive/Mathematical_BPE_326_Patients.csv'
df = pd.read_csv(csv_path)

# 2. Continuous Score의 분위수(Quantile) 계산 (데이터 분포 파악)
q25 = df['Math_BPE_Score_Continuous'].quantile(0.25)
q50 = df['Math_BPE_Score_Continuous'].quantile(0.50)
q75 = df['Math_BPE_Score_Continuous'].quantile(0.75)

print(f"📊 연속형 점수 분위수 기준점:")
print(f" - 하위 25% (Minimal -> Mild): {q25:.2f}점")
print(f" - 하위 50% (Mild -> Moderate): {q50:.2f}점")
print(f" - 하위 75% (Moderate -> Marked): {q75:.2f}점")

# 3. 분위수 기반 상대적 4단계 Class 지정 함수
def assign_quantile_class(score):
    if score < q25:
        return 0  # Minimal (하위 25%)
    elif score < q50:
        return 1  # Mild (25% ~ 50%)
    elif score < q75:
        return 2  # Moderate (50% ~ 75%)
    else:
        return 3  # Marked (상위 25%)

# 4. Class 재할당
df['Math_BPE_Class_4'] = df['Math_BPE_Score_Continuous'].apply(assign_quantile_class)

# 5. 결과 분포 확인
print("\n🎉 재조정된 Math_BPE_Class_4 데이터 분포:")
print(df['Math_BPE_Class_4'].value_counts().sort_index())

# 6. 보정된 CSV 및 엑셀 덮어쓰기 저장
df.to_csv(csv_path, index=False, encoding='utf-8-sig')
df.to_excel('/content/drive/MyDrive/Mathematical_BPE_326_Patients.xlsx', index=False, engine='openpyxl')

print(f"\n✅ 보정 완료! 파일이 새로 업데이트되었습니다.")
```

```python
import pandas as pd
import numpy as np
import os
from tkinter import Tk, filedialog

# 1. 파일 선택 창 띄우기 (연속형 점수인 'Math_BPE_Score_Continuous'가 포함된 원본 CSV 선택)
root = Tk()
root.withdraw()
root.attributes('-topmost', True)

print("📂 'Math_BPE_Score_Continuous' 컬럼이 포함된 원본 CSV 파일을 선택해주세요!")
csv_path = filedialog.askopenfilename(
    title="원본 수학적 정량화 CSV 파일 선택",
    filetypes=[("CSV Files", "*.csv"), ("All Files", "*.*")]
)

if not csv_path:
    print("❌ 파일이 선택되지 않았습니다.")
else:
    print(f"✅ 선택된 파일 경로: {csv_path}")
    df = pd.read_csv(csv_path)

    # 2. 연속형 점수 컬럼 존재 여부 확인
    score_col = None
    for candidate in ['Math_BPE_Score_Continuous', 'Score', 'Continuous_Score']:
        if candidate in df.columns:
            score_col = candidate
            break

    if score_col is None:
        print(f"❌ 에러: 연속형 점수 컬럼을 찾을 수 없습니다. 현재 컬럼 목록: {df.columns.tolist()}")
    else:
        print(f"🔍 감지된 연속형 점수 컬럼명: '{score_col}'")

        # 3. 전문의 판독 비율 통계 반영 (Minimal: 40%, Mild: 35%, Moderate: 20%, Marked: 5%)
        p40 = df[score_col].quantile(0.40)
        p75 = df[score_col].quantile(0.75)
        p95 = df[score_col].quantile(0.95)

        print(f"\n📊 전문의 비율 통계(40% / 35% / 20% / 5%) 반영 기준점:")
        print(f" - Minimal -> Mild 경계 (하위 40%): {p40:.2f}점")
        print(f" - Mild -> Moderate 경계 (하위 75%): {p75:.2f}점")
        print(f" - Moderate -> Marked 경계 (하위 95%): {p95:.2f}점")

        # 4. 임상 판독 비율에 맞춘 다단계 Class 지정 함수
        def assign_clinical_ratio_class(score):
            if score < p40:
                return 0  # Minimal (~40%)
            elif score < p75:
                return 1  # Mild (40% ~ 75%)
            elif score < p95:
                return 2  # Moderate (75% ~ 95%)
            else:
                return 3  # Marked (95% ~ 100%, 상위 5%)

        # 5. Class 재할당
        df['Math_BPE_Class_4'] = df[score_col].apply(assign_clinical_ratio_class)

        # 6. 결과 분포 확인
        print("\n🎉 임상 비율 반영 후 Math_BPE_Class_4 데이터 분포:")
        class_counts = df['Math_BPE_Class_4'].value_counts().sort_index()
        print(class_counts)
        print("\n📈 비율(%):")
        print((class_counts / len(df) * 100).round(2))

        # 7. 학습용 포맷에 맞춰 정리 (Filename, Math_BPE)
        df['Math_BPE'] = df['Math_BPE_Class_4']
        df_clean = df[['Filename', 'Math_BPE']].copy()
        
        # 파일명 문자열 정돈
        df_clean['Filename'] = df_clean['Filename'].str.replace('_CM', '', regex=False)
        df_clean['Filename'] = df_clean['Filename'].str.replace('_padded.png', '', regex=False)
        df_clean['Filename'] = df_clean['Filename'].str.replace('.png', '', regex=False)

        # 저장 경로 설정
        directory = os.path.dirname(csv_path)
        clean_csv_save_path = os.path.join(directory, 'Mathematical_BPE_Clean_Labels.csv')

        df.to_csv(csv_path, index=False, encoding='utf-8-sig')
        df_clean.to_csv(clean_csv_save_path, index=False, encoding='utf-8-sig')

        print(f"\n✅ 임상 비율 통계 기반 클래스 재조정 완료!")
        print(f"📁 저장된 Clean Labels 경로: {clean_csv_save_path}")
```
